```python
#dividing data into male and female
df_m = df[df['Gender (1 = female 2 = male)'] == 2]
df_m = df_m.drop(columns=['Gender (1 = female 2 = male)'])

df_f = df[df['Gender (1 = female 2 = male)'] == 1]
df_f = df_f.drop(columns=['Gender (1 = female 2 = male)'])
```

```python
df_m_checks = df_m.drop(columns=['Age','Day1HoursUse', 'Day2HoursUse',
       'Day3HoursUse', 'Day4HoursUse', 'Day5HoursUse', 'Day6HoursUse',
       'Day7HoursUse', 'Day8HoursUse', 'Day9HoursUse', 'Day10HoursUse',
       'Day11HoursUse', 'Day12HoursUse', 'Day13HoursUse', 'totalhoursweek1',
       'totalhoursweek2', 'total_use'])
df_f_checks = df_f.drop(columns=['Age','Day1HoursUse', 'Day2HoursUse',
       'Day3HoursUse', 'Day4HoursUse', 'Day5HoursUse', 'Day6HoursUse',
       'Day7HoursUse', 'Day8HoursUse', 'Day9HoursUse', 'Day10HoursUse',
       'Day11HoursUse', 'Day12HoursUse', 'Day13HoursUse', 'totalhoursweek1',
       'totalhoursweek2', 'total_use'])
```

```python
df_m_use = df_m.drop(columns=['Age','Day1TotalChecks', 'Day2TotalChecks',
       'Day3TotalChecks', 'Day4TotalChecks', 'Day5TotalChecks',
       'Day6TotalChecks', 'Day7TotalChecks', 'Day8TotalChecks',
       'Day9TotalChecks', 'Day10TotalChecks', 'Day11TotalChecks',
       'Day12TotalChecks', 'Day13TotalChecks','totalchecksweek1', 'totalchecksweek2','total_checks'])
df_f_use = df_f.drop(columns=['Age','Day1TotalChecks', 'Day2TotalChecks',
       'Day3TotalChecks', 'Day4TotalChecks', 'Day5TotalChecks',
       'Day6TotalChecks', 'Day7TotalChecks', 'Day8TotalChecks',
       'Day9TotalChecks', 'Day10TotalChecks', 'Day11TotalChecks',
       'Day12TotalChecks', 'Day13TotalChecks','totalchecksweek1', 'totalchecksweek2','total_checks'])
```

```python
fig, ax =plt.subplots(1,2,figsize=(20,8))

df1 = df_f_use.drop(['Participant ID'], axis=1)
sns.heatmap(df1.corr(),cmap='Greens', ax=ax[0])

df2 = df_m_use.drop(['Participant ID'], axis=1)
sns.heatmap(df2.corr(),cmap='Greens', ax=ax[1])
print('\n{:>40}'.format('Female hour use'),'{:>66}'.format('Male hour use'))
```

```python
fig, ax =plt.subplots(1,2,figsize=(20,8))

df1 = df_f_checks.drop(['Participant ID'], axis=1)
sns.heatmap(df1.corr(),cmap='Greens', ax=ax[0])

df2 = df_m_checks.drop(['Participant ID'], axis=1)
sns.heatmap(df2.corr(),cmap='Greens', ax=ax[1])
print('\n{:>40}'.format('Female checks'),'{:>66}'.format('Male checks'))
```

```python
fig, ax =plt.subplots(1,2,figsize=(12,4))
sns.boxplot(x=df_f_use['total_use'], ax=ax[0])
sns.boxplot(x=df_m_use['total_use'], ax=ax[1])
print('\n{:>25}'.format('Female total hour use'),'{:>51}'.format('Male total hour use'))
print('    Median ',df_f_use['total_use'].median(),'                                     Median ',df_m_use['total_use'].median())
#print('\n{:>20}'.format(df_f_use.median()),'{:>46}'.format(df_m_use.median()))
```
