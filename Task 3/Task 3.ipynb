```python
# Calculating mean score to figure out unwanted entries
df_new.columns
columns = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10', 'Q11',
       'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'Q20', 'Q21',
       'Q22', 'Q23', 'Q24', 'Q25', 'Q26', 'Q27', 'Q28', 'Q29', 'Q30', 'Q31',
       'Q32', 'Q33', 'Q34', 'Q35', 'Q36']
df_new['mean_score'] = df_new[list(columns)].sum(axis=1)/36
df_new
```

```python
# Filtering people out to have realistic dataset
# saturating data by removing like people who filled all answers as 0 or 5
df1 = df_new[(df_new['mean_score'] < 4.5) & (df_new['mean_score'] > 1.5)]
pd.set_option('display.max_columns', 50)
df1 = df1.reset_index()
df1 = df1.drop(columns=['index'])
df1.head()
```

```python
sns.kdeplot(data=df1['mean_score'], shade=True)
```

```python
# scoring scheme

# We'll club the questions which are similar
# we'll find standard deviation of point marked by candidate for those common questions
# print out the standard deviation as score for that particular set of ques
# after doing this for all sets we'll see the 
```

```python
#clubbing Q1 & Q15 
for i in range(len(df1)):
  A = 5 - df1.at[i,'Q1']
  B = df1.at[i,'Q15']
  sd = np.std([A,B])
  df1.at[i,'sd_q1_q15'] = sd
```

```python
#clubbing Q2 & Q14 & Q22
for i in range(len(df1)):
  A = df1.at[i,'Q2']
  B = df1.at[i,'Q14']
  C = 5 - df1.at[i,'Q22']
  sd = np.std([A,B,C])
  df1.at[i,'sd_q2_q14_q22'] = sd
```

```python
df2 = df1[(df1.sd_q1_q15 <= 1) & (df1.sd_q2_q14_q22 <= 1) & (df1.sd_q3_q7_q23 <= 1) & (df1.sd_q5_q11 <= 1) & (df1.sd_q6_q10 <= 1)]
df2 = df2[(df2.sd_q13_q17_q19 <= 1) & (df2.sd_q21_q29 <= 1) & (df2.sd_q25_q27 <= 1) & (df2.sd_q30_q36 <= 1) & (df2.sd_q31_q33 <= 1)]
df2
```

```python
columns = ['sd_q1_q15', 'sd_q2_q14_q22', 'sd_q3_q7_q23', 'sd_q5_q11', 'sd_q6_q10', 'sd_q13_q17_q19', 'sd_q21_q29', 'sd_q25_q27', 'sd_q31_q33', 'sd_q30_q36']
df2['sd_score'] = df2[list(columns)].sum(axis=1)
df2
```

```python
sns.kdeplot(data=df2['sd_score'], shade=True)
```

```python
mean = df2['sd_score'].mean()
std = df2['sd_score'].std()

print('Mean of sd_score              ',mean)
print('Standard Deviation of sd_score',std)
```

```python
# Considering the above plot to be very close to the normal distribution curve 
# Z=(x-μ)/σ.
# z = z_score, μ=mean, σ=standard deviation, x= value at i^th percentile of normal distribution
# x = μ + Z σ

# We can find Z_score by using the z score table
# By intution considering only 90 percentile of the people to reduce the inconsistency

# z_score for 90% probability from z score table is 1.28
```

```python
z=1.28
X = mean + z*std
print('sd_score value corresponding to the 90th percentile for this distribution is :',X)
```

```python
df3 = df2[(df2['sd_score'] < X)]
df3.reset_index(inplace = True)
df3.drop(columns='index')
N1 = df3[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26', 'Q27', 'Q28', 'Q29', 'Q30', 'Q31', 'Q32', 'Q33', 'Q34', 'Q35', 'Q36', 'age', 'gender', 'country']]
N1.to_csv('N1.csv')
print("")
print("N1".center(120))
N1
```
